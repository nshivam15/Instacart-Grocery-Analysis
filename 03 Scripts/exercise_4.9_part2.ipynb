#### Task 3

```python
#create histogram
hist = df_merged['order_hour_of_day'].plot.hist(bins = 24)

#add title and axis label
hist.set_title('Frequency of orders each hour of the day')
hist.set_xlabel('Hour of Day')
```

#### Task 4

```python
#create a bar chart for loyalty_flag
bar = df_merged['loyalty_flag'].value_counts().plot.bar()
bar.set_title("Distribution of orders for loyal customers")
bar.set_xlabel("Loyalty Flag")
bar.set_ylabel("Number of orders")
```

#### Task 5

```python
#Create a random sample 
np.random.seed(4)
```

```python
#create a list of holding True/False values to the test np.ransom.rand()<=0.7
dev = np.random.rand(len(df_merged)) <= 0.7
```

```python
#store 70% of the sample in the data frame big
big = df_merged[dev]
```

```python
#store 30% or the sample in the dataframe small
small = df_merged[~dev]
```

```python
#check the len of acutal df
len(df_merged)
```

```python
#check the lengeth of split df
len(big) + len(small)
```

```python
# A new dataframe that contains only the "prices" and "order_hour_of_day" columns from small subset
df_2 = small[['order_hour_of_day', 'prices']]
#check the output
df_2.head()
```

```python
#create line chart
line1 = sns.lineplot(data = df_2, x = 'order_hour_of_day' , y = 'prices')
```

#### Task 6

```python
#new dataframe that contains only the "age" and "number_of_dependants" columns from small subset
df_3 = small.loc[:, ['age', 'num_of_dependents']]
```

```python
#create line chart
line2 = sns.lineplot(data = df_3, x = 'age' , y = 'num_of_dependents')
```

There is no clear correlation between age and number of dependents.

#### Task 7

```python
#create new df
df_4 = small[['age','income']]
```

```python
#create scatterplot
scatterplot = sns.scatterplot(x= 'age', y = 'income', data = df_4)
```

The scatterplot displays age on the x-axis and income on the y-axis. Points are widely scattered with no clear upward or downward trend, suggesting that income is not strongly related to age in this dataset. in conclusion, the plot suggests income varies greatly at every age and age alone doesnt predict higher or lower earnings in this sample.
