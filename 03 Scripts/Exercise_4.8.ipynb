###  Grouping Data & Aggregating Variables

```python
# Importing Libraries
import pandas as pd
import numpy as np
import os
```

```python
#Creating string of project folder.

path = r'/Users/nautis1/data_project/09-2025-Instacart-Basket-Analysis/Instacart Basket Analysis'
```

```python
#importing orders_products_combined dataframe from the pickle file
orders_products_combined = pd.read_pickle(os.path.join(path, '02 data' , 'prepared data' , 'ords_prods_merge_new_var.pkl'))
```

```python
#cleaning prices column 
orders_products_combined.loc[orders_products_combined['prices'] >100, 'prices'] = np.nan
```

#### Task 4

```python
# create max_order column
orders_products_combined['max_order'] = orders_products_combined.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
orders_products_combined.head()
```

```python
#create loyalty_flag
orders_products_combined.loc[orders_products_combined['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#create loyalty_flag
orders_products_combined.loc[(orders_products_combined['max_order'] <= 40) & (orders_products_combined['max_order'] > 10) , 'loyalty_flag'] = 'Regular customer'
```

```python
#create loyalty_flag
orders_products_combined.loc[orders_products_combined['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
orders_products_combined['loyalty_flag'].value_counts(dropna = False)
```

#### Task 5

```python
# get descriptive statistics for customer spending across categories
orders_products_combined.groupby(['loyalty_flag']).agg({'prices': ['mean', 'min', 'max']})
```

The prices statistics reveals opposite pattern than number of orders meaning loyal customers have most number of orders but lowest mean which might be due to the number of orders they place the mean gets on the lower side where as for regular customers since the number of orders are less an outlier on the higher side of price will spike the mean considerably. 

#### Task 6

```python
#find the avg price across all the orders by the user
orders_products_combined['avg_price'] = orders_products_combined.groupby(['user_id'])['prices'].transform(np.average)
```

```python
orders_products_combined.head()
```

```python
#create a spending flag for low spender
orders_products_combined.loc[orders_products_combined['avg_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
#populate spending flag for high spender
orders_products_combined.loc[orders_products_combined['avg_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
orders_products_combined['spending_flag'].value_counts(dropna= False)
```

#### Task 7

```python
#find the mediam days_since_prior order for all users
orders_products_combined['median_days'] = orders_products_combined.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#create regularity flag 
orders_products_combined.loc[orders_products_combined['median_days'] > 20, 'regularity_flag'] = 'Non-frequent customer'
```

```python
#create regularity flag for regular customer
orders_products_combined.loc[(orders_products_combined['median_days'] > 10) & (orders_products_combined['median_days'] <=20), 'regularity_flag'] = 'Regular customer'
```

```python
#create flag for Frequent customer
orders_products_combined.loc[orders_products_combined['median_days'] <= 10, 'regularity_flag'] = 'Frequent Customer'
```

```python
orders_products_combined['regularity_flag'].value_counts(dropna= False)
```
